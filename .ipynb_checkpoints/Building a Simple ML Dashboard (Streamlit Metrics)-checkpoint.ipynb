**PYTHON SCRIPT TO _BUILD A SIMPLE STREAMLIT DASHBOARD FOR MONITORING ML MODEL METRICS_. 🧠📊🚀**

This script demonstrates how to create a simple **ML monitoring dashboard** using Streamlit. The dashboard displays model metrics such as:

- accuracy
- precision
- recall
- F1-score
- prediction volume
- average confidence

---

## **📦 Import Required Libraries**

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
```

---

## **📝 Create Sample Monitoring Data**

```python
monitoring_data = pd.DataFrame({
    "date": pd.date_range(
        start="2025-01-01",
        periods=10,
        freq="D"
    ),

    "accuracy": [
        0.96, 0.95, 0.95, 0.94, 0.93,
        0.92, 0.91, 0.90, 0.89, 0.88
    ],

    "precision": [
        0.95, 0.94, 0.94, 0.93, 0.92,
        0.91, 0.90, 0.89, 0.88, 0.87
    ],

    "recall": [
        0.94, 0.94, 0.93, 0.92, 0.91,
        0.90, 0.89, 0.88, 0.87, 0.86
    ],

    "f1_score": [
        0.945, 0.94, 0.935, 0.925, 0.915,
        0.905, 0.895, 0.885, 0.875, 0.865
    ],

    "prediction_volume": [
        500, 520, 510, 540, 560,
        580, 600, 610, 630, 650
    ],

    "avg_confidence": [
        0.94, 0.94, 0.93, 0.92, 0.91,
        0.90, 0.89, 0.88, 0.87, 0.86
    ]
})
```

---

## **🖥️ Create Dashboard Title**

```python
st.title("ML Model Monitoring Dashboard")
st.write("Simple dashboard to track model performance over time.")
```

---

## **📊 Display Latest Metrics**

```python
latest = monitoring_data.iloc[-1]

col1, col2, col3, col4 = st.columns(4)

col1.metric("Accuracy", round(latest["accuracy"], 3))
col2.metric("Precision", round(latest["precision"], 3))
col3.metric("Recall", round(latest["recall"], 3))
col4.metric("F1 Score", round(latest["f1_score"], 3))
```

---

## **📈 Plot Accuracy Trend**

```python
st.subheader("Accuracy Trend")

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(
    monitoring_data["date"],
    monitoring_data["accuracy"],
    marker="o"
)

ax.set_xlabel("Date")
ax.set_ylabel("Accuracy")
ax.set_title("Model Accuracy Over Time")
ax.grid(True)

st.pyplot(fig)
```

---

## **📈 Plot Prediction Volume**

```python
st.subheader("Prediction Volume Trend")

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(
    monitoring_data["date"],
    monitoring_data["prediction_volume"],
    marker="o"
)

ax.set_xlabel("Date")
ax.set_ylabel("Prediction Count")
ax.set_title("Prediction Volume Over Time")
ax.grid(True)

st.pyplot(fig)
```

---

## **🚨 Add Simple Alert Logic**

```python
st.subheader("Monitoring Alerts")

if latest["accuracy"] < 0.90:
    st.warning("Accuracy has dropped below threshold.")
else:
    st.success("Accuracy is within healthy range.")

if latest["avg_confidence"] < 0.87:
    st.warning("Average confidence is decreasing.")
else:
    st.success("Average confidence is within healthy range.")
```

---

## **📋 Show Raw Monitoring Data**

```python
st.subheader("Monitoring Data")
st.dataframe(monitoring_data)
```
