**PYTHON SCRIPT TO _BUILD A PRODUCTION-READY PREPROCESSING PIPELINE FOR MIXED NUMERIC AND CATEGORICAL DATA USING COLUMNTRANSFORMER_. 🧠🔧📊**

This script demonstrates how to use **ColumnTransformer** to apply different preprocessing steps to **numeric** and **categorical** columns in the same dataset. This is a critical technique for real-world ML systems where data is rarely purely numeric.

---

## **📦 Import Required Libraries**

We’ll use pandas for tabular data and scikit-learn for preprocessing, pipeline creation, and modeling.

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

---

## **🧩 Create a Mixed-Type Dataset**

We create a small sample dataset containing both numeric and categorical features.

```python
df = pd.DataFrame({
    "age": [25, 32, 47, 51, 62, 23, 36, 44],
    "income": [30000, 50000, 70000, 65000, 90000, 28000, 52000, 61000],
    "gender": ["Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male"],
    "city": ["A", "B", "A", "C", "B", "A", "C", "B"],
    "purchased": [0, 1, 1, 0, 1, 0, 1, 0]
})
```

This mirrors real-world datasets where customer, business, or operational data includes mixed feature types.

---

## **✂️ Split Features and Target**

We separate the target column and split the data into training and testing sets.

```python
X = df.drop("purchased", axis=1)
y = df["purchased"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

---

## **🔢 Define Numeric and Categorical Columns**

We explicitly specify which columns are numeric and which are categorical.

```python
numeric_features = ["age", "income"]
categorical_features = ["gender", "city"]
```

This separation is what enables ColumnTransformer to apply the correct preprocessing to each group.

---

## **🔧 Build Preprocessing Pipelines**

We define separate preprocessing steps for numeric and categorical data.

```python
numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

- Numeric columns → missing value handling + scaling
- Categorical columns → missing value handling + one-hot encoding

---

## **🧱 Combine with ColumnTransformer**

We now combine both preprocessing pipelines into a single ColumnTransformer.

```python
preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

This ensures that each column type is handled correctly and consistently.

---

## **🤖 Build Full Pipeline with Model**

We combine preprocessing and model training into one end-to-end workflow.

```python
model_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(n_estimators=200, random_state=42))
])
```

This is the production-safe approach — the same preprocessing will be applied at both training and inference time.

---

## **🚀 Train the Pipeline**

```python
model_pipeline.fit(X_train, y_train)
```

All numeric and categorical preprocessing now happens automatically inside the pipeline.

---

## **📊 Evaluate the Model**

We generate predictions and evaluate the pipeline’s performance.

```python
y_pred = model_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

---

## **🔍 Predict on New Mixed-Type Data**

We can now pass raw mixed-type data directly into the pipeline.

```python
new_data = pd.DataFrame({
    "age": [40, 29],
    "income": [58000, 32000],
    "gender": ["Female", "Male"],
    "city": ["C", "A"]
})

predictions = model_pipeline.predict(new_data)
print("Predictions:", predictions)
```

No manual encoding or scaling is needed — the pipeline handles everything.

---

## **🧠 Key Takeaways**

1. Real-world ML datasets often contain both numeric and categorical features.
2. ColumnTransformer allows different preprocessing for each feature type.
3. It helps keep preprocessing clean, structured, and production-ready.
4. Combining ColumnTransformer with Pipeline prevents preprocessing mistakes.
5. A foundational concept for scalable, real-world machine learning systems.
