**PYTHON SCRIPT TO _HANDLE HIGH-CARDINALITY CATEGORICAL FEATURES FOR MACHINE LEARNING_. 🧠🏷️📊**

This script demonstrates how to encode categorical features with many unique values (high cardinality) using practical techniques suitable for tabular ML pipelines.

---

## **📦 Import Required Libraries**

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

---

## **🧩 Create Sample Dataset**

We simulate a dataset with a high-cardinality categorical column.

```python
np.random.seed(42)

n_samples = 1000

data = pd.DataFrame({
    "user_id": [f"user_{i}" for i in range(n_samples)],
    "city": np.random.choice(
        [f"city_{i}" for i in range(200)],
        size=n_samples
    ),
    "age": np.random.randint(18, 60, size=n_samples),
    "purchased": np.random.randint(0, 2, size=n_samples)
})
```

The `city` column contains many unique categories.

---

## **🚨 Problem with One-Hot Encoding**

Applying one-hot encoding directly:

```python
pd.get_dummies(data["city"])
```

can create:

- hundreds or thousands of columns
- sparse datasets
- high memory usage
- slower model training

---

## **✅ Frequency Encoding**

A practical alternative is frequency encoding.

```python
city_freq = data["city"].value_counts()

data["city_freq_encoded"] = data["city"].map(city_freq)
```

Each category is replaced by how frequently it appears.

---

## **✂️ Prepare Features**

```python
X = data[["age", "city_freq_encoded"]]
y = data["purchased"]
```

---

## **✂️ Split Data**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

---

## **🤖 Train Model**

```python
model = LogisticRegression(max_iter=5000)

model.fit(X_train, y_train)
```

---

## **📊 Evaluate Model**

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

---

## **🔍 Why High-Cardinality Encoding Matters**

High-cardinality features are common in:

- user IDs
- product IDs
- city names
- search terms
- transaction categories

Naive encoding methods can explode feature dimensions.

Efficient encoding improves:

- memory usage
- training speed
- model scalability

---

## **🧠 Other Common Techniques**

Besides frequency encoding:

- Target Encoding
- Hash Encoding
- Embeddings (Deep Learning)
- Leave-One-Out Encoding

are also widely used.

---

## **🧠 Key Takeaways**

1. High-cardinality features contain many unique categories.
2. One-hot encoding can become inefficient at scale.
3. Frequency encoding is a lightweight alternative.
4. Efficient encoding improves scalability and training speed.
5. Handling categorical data correctly is critical in tabular ML.

---

## **Conclusion**

Encoding high-cardinality categorical features is an important challenge in real-world machine learning systems. By using techniques like frequency encoding, we can efficiently represent large categorical spaces while keeping models scalable and practical.
