**PYTHON SCRIPT TO _HANDLE UNSEEN CATEGORICAL VALUES SAFELY IN PRODUCTION MACHINE LEARNING PIPELINES_. 🧠🏭🔤**

This script demonstrates how to build a production-safe preprocessing pipeline that can handle **new categorical values** at inference time — values that were never seen during training. This is a common real-world problem in deployed ML systems.

---

## **📦Import Required Libraries**

We’ll use pandas for tabular data and scikit-learn for preprocessing, pipeline creation, and modeling.

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

---

## **🧩 Create a Mixed-Type Dataset**

We create a small sample dataset with numeric and categorical columns.

```python
df = pd.DataFrame({
    "age": [25, 32, 47, 51, 62, 23, 36, 44],
    "income": [30000, 50000, 70000, 65000, 90000, 28000, 52000, 61000],
    "gender": ["Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male"],
    "city": ["A", "B", "A", "C", "B", "A", "C", "B"],
    "purchased": [0, 1, 1, 0, 1, 0, 1, 0]
})
```

This mirrors real-world customer or user data where categorical features are common.

---

## **✂️ Split Features and Target**

We separate input features and target labels.

```python
X = df.drop("purchased", axis=1)
y = df["purchased"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

---

## **🔢 Define Numeric and Categorical Columns**

We explicitly identify which columns need numeric preprocessing and which need categorical encoding.

```python
numeric_features = ["age", "income"]
categorical_features = ["gender", "city"]
```

---

## **🔧 Build Preprocessing Pipelines**

The important part here is handle_unknown="ignore" inside OneHotEncoder.

```python
numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

This ensures that if a new category appears in production, the pipeline does **not crash**.

---

## **🧱 Combine with ColumnTransformer**

We merge numeric and categorical preprocessing into one unified step.

```python
preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

---

## **🤖 Build the Full Pipeline**

We combine preprocessing and modeling into one end-to-end production-safe workflow.

```python
model_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(n_estimators=200, random_state=42))
])
```

---

## **🚀 Train the Pipeline**

```python
model_pipeline.fit(X_train, y_train)
```

At this point, the encoder learns categories seen in the training data.

---

## **📊 Evaluate the Model**

We first verify that the normal evaluation flow works correctly.

```python
y_pred = model_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

---

## **🔍 Test Unseen Categories in Production**

Now we simulate real production data containing **new categories** never seen during training.

```python
new_data = pd.DataFrame({
    "age": [40, 29],
    "income": [58000, 32000],
    "gender": ["Other", "Male"],   # "Other" was not seen during training
    "city": ["D", "A"]             # "D" was not seen during training
})
```

Without handle_unknown="ignore", this would raise an error.

---

## **✅ Predict Safely on New Data**

Because we configured the encoder correctly, predictions still work.

```python
predictions = model_pipeline.predict(new_data)
print("Predictions:", predictions)
```

The unseen categories are ignored safely, allowing the system to continue serving predictions.

---

## **🔍 Why This Matters in Production**

In deployed systems, new categories appear all the time:

- new cities
- new user types
- new product types
- new payment methods

If the pipeline is not configured properly, inference breaks.

Handling unseen categories safely is one of the most important **production ML safeguards**.

----

## **🧠 Key Takeaways**

1. Unseen categories are a common production ML problem.
2. OneHotEncoder(handle_unknown="ignore") prevents inference failures.
3. Pipelines keep preprocessing and modeling consistent.
4. Mixed-type production data must be handled safely and automatically.
5. This is a critical technique for reliable real-world ML systems.
