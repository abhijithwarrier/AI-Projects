**PYTHON SCRIPT TO _COMPARE SAVING ONLY THE MODEL VS SAVING THE FULL PREPROCESSING PIPELINE FOR PRODUCTION ML SYSTEMS_. 🧠💾⚖️**

This script demonstrates a critical production ML concept:

Why saving **only the model** can break your system, and why saving the **full pipeline** is the correct approach.

---

## **📦 Import Required Libraries**

We use preprocessing tools, pipelines, and joblib for saving models.

```python
import joblib
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

---

## **🧩 Load Dataset**

```python
data = load_wine()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🚨 ❌ Approach 1 – Saving Only the Model**

Here we train using scaled data but save only the model.

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

model = LogisticRegression(max_iter=5000)
model.fit(X_train_scaled, y_train)

# Save ONLY model
joblib.dump(model, "model_only.pkl")
```

---

## **⚠️ Problem During Inference**

When we load the model later, we don’t have the scaler.

```python
loaded_model = joblib.load("model_only.pkl")

# ❌ Missing preprocessing step
y_pred_wrong = loaded_model.predict(X_test)

print("❌ Accuracy (model only):", accuracy_score(y_test, y_pred_wrong))
```

👉 Problem:

- Model expects scaled input
- Raw data is passed
- Predictions become incorrect

---

## **✅ Approach 2 – Saving Full Pipeline**

We combine preprocessing + model into a pipeline.

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=5000))
])

pipeline.fit(X_train, y_train)
```

---

## **💾 Save Full Pipeline**

```python
joblib.dump(pipeline, "full_pipeline.pkl")
```

Now everything needed for inference is stored together.

---

## **🚀 Load and Use Pipeline**

```python
loaded_pipeline = joblib.load("full_pipeline.pkl")

y_pred_correct = loaded_pipeline.predict(X_test)

print("✅ Accuracy (full pipeline):", accuracy_score(y_test, y_pred_correct))
```

👉 No preprocessing mistakes

👉 Fully consistent predictions

---

## **🔍 Why Full Pipeline Saving Matters**

Saving only the model:

- ❌ loses preprocessing logic
- ❌ causes inference errors
- ❌ breaks production systems

Saving full pipeline:

- ✅ keeps transformations intact
- ✅ ensures consistency
- ✅ simplifies deployment
- ✅ reduces human error

---

## **🧠 Key Takeaways**

1. Saving only the model is not enough in production ML.
2. Preprocessing steps must be preserved.
3. Pipelines bundle preprocessing + model together.
4. Saving the full pipeline ensures consistent inference.
5. A critical best practice for real-world ML systems.

---

## **Conclusion**

In production machine learning, saving only the trained model is a common mistake that leads to inconsistent and unreliable predictions. By saving the full pipeline, we ensure that preprocessing and modeling steps are always applied together, making the system robust and production-ready.

This is a key concept in the **Production ML** track of the **AI with Python** series — focusing on building reliable and deployable ML systems.
