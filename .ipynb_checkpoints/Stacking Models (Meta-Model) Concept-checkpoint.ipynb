**PYTHON SCRIPT TO _COMBINE MULTIPLE MACHINE LEARNING MODELS USING STACKING ENSEMBLES_. 🧠🧩🚀**

This script demonstrates the concept of **stacking**, where predictions from multiple base models are combined using a **meta-model** to improve overall performance.

---

## **📦 Import Required Libraries**

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
```

---

## **🧩 Load Dataset**

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🤖 Define Base Models**

We use multiple different algorithms as base learners.

```python
base_models = [
    ("rf", RandomForestClassifier(random_state=42)),
    ("svc", SVC(probability=True, random_state=42)),
    ("lr", LogisticRegression(max_iter=5000))
]
```

Each model learns different patterns from the data.

---

## **🧠 Define the Meta-Model**

The meta-model learns from predictions generated by base models.

```python
meta_model = LogisticRegression(max_iter=5000)
```

---

## **🚀 Build the Stacking Classifier**

```python
stacking_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model
)
```

This combines all base model predictions into a final prediction.

---

## **🤖 Train the Stacking Model**

```python
stacking_model.fit(X_train, y_train)
```

---

## **📊 Generate Predictions**

```python
y_pred = stacking_model.predict(X_test)
```

---

## **✅ Evaluate Model Performance**

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

---

## **🔍 How Stacking Works**

Stacking follows a layered approach:

### **🔹 Base Models**

- learn independently
- generate predictions

### **🔹 Meta-Model**

- receives predictions from base models
- learns how to combine them optimally

This often improves generalization and predictive power.

---

## **🧠 Why Stacking Is Powerful**

Different models capture different patterns:

- RandomForest → strong non-linear learning
- SVM → margin-based separation
- Logistic Regression → linear decision boundaries

Stacking combines their strengths into one system.

---

## **🧠 Key Takeaways**

1. Stacking combines multiple ML models into an ensemble.
2. Base models learn independently from the same dataset.
3. A meta-model learns from base model predictions.
4. Stacking often improves performance and generalization.
5. A powerful ensemble strategy in advanced ML systems.

---

## **Conclusion**

Stacking is an advanced ensemble technique that combines multiple models to create stronger predictive systems. By leveraging the strengths of different algorithms, stacking often produces more robust and accurate results than individual models alone.
