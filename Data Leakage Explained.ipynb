**PYTHON SCRIPT TO _IDENTIFY AND PREVENT DATA LEAKAGE IN MACHINE LEARNING WORKFLOWS_. 🧠🚨🔍**

This script demonstrates what **data leakage** is, how it silently inflates model performance, and how to correctly prevent it using proper training pipelines.

---

## **📦 Import Required Libraries**

We use scikit-learn utilities for preprocessing, modeling, and evaluation.

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

---

## **🧩 Load Dataset**

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **🚨 ❌ WRONG APPROACH – Data Leakage Example**

Here we scale the entire dataset **before splitting**, which leaks information from test data into training.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   # ❌ fitted on full dataset

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("❌ Accuracy with Data Leakage:", accuracy_score(y_test, y_pred))
```

👉 Problem:

- The scaler learned statistics from **both train and test data**
- This makes the model unrealistically strong

---

## **✅ CORRECT APPROACH – No Data Leakage**

We split first, then apply preprocessing only on training data.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)   # ✅ fit only on train
X_test_scaled = scaler.transform(X_test)         # ✅ transform test only

model = LogisticRegression(max_iter=5000)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("✅ Accuracy without Leakage:", accuracy_score(y_test, y_pred))
```

---

## **🔧 BEST PRACTICE – Using Pipeline**

The safest way to prevent leakage is using a pipeline.

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=5000))
])

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)

print("🚀 Pipeline Accuracy (Leak-Free):", accuracy_score(y_test, y_pred))
```

Pipelines ensure preprocessing is applied correctly and consistently.

---

## **🔍 Why Data Leakage Is Dangerous**

Data leakage leads to:

- overly optimistic accuracy
- poor real-world performance
- incorrect business decisions
- loss of trust in ML systems

It is one of the most common mistakes in machine learning.

---

## **🧠 Key Takeaways**

1. Data leakage occurs when test data influences training.
2. Preprocessing before splitting is a common mistake.
3. Always split data before applying transformations.
4. Pipelines help prevent leakage automatically.
5. A critical concept for building reliable ML systems.

---

## **Conclusion**

Data leakage is a silent but dangerous problem in machine learning. It can make models appear highly accurate during development but fail in production. By following correct workflows and using pipelines, we can ensure that our models remain trustworthy and robust.
