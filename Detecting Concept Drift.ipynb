**PYTHON SCRIPT TO _DETECT CONCEPT DRIFT BY MONITORING CHANGES IN MODEL PERFORMANCE OVER TIME_. 🧠📉🚨**

This script demonstrates a simplified approach to detecting **concept drift**, where the relationship between features and the target variable changes over time, causing model performance to degrade. Concept drift is one of the most important challenges in production machine learning systems.

---

## **📦 Import Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

---

## **📝 Simulate Model Performance Over Time**

We create historical performance metrics collected after deployment.

```python
performance_data = pd.DataFrame({
    "date": pd.date_range(
        start="2025-01-01",
        periods=12,
        freq="ME"
    ),

    "accuracy": [
        0.95, 0.95, 0.94, 0.94,
        0.93, 0.92, 0.91, 0.89,
        0.87, 0.85, 0.83, 0.80
    ]
})
```

---

## **🔍 View Performance Data**

```python
print(performance_data)
```

---

## **📈 Visualize Accuracy Trend**

```python
plt.figure(figsize=(8, 4))

plt.plot(
    performance_data["date"],
    performance_data["accuracy"],
    marker="o"
)

plt.title("Model Accuracy Over Time")
plt.xlabel("Date")
plt.ylabel("Accuracy")

plt.grid(True)
plt.tight_layout()
plt.show()
```

---

## **📊 Measure Performance Drop**

Compare current performance against historical baseline.

```python
baseline_accuracy = performance_data["accuracy"].iloc[0]

latest_accuracy = performance_data["accuracy"].iloc[-1]

accuracy_drop = (
    baseline_accuracy - latest_accuracy
)

print(
    "Accuracy Drop:",
    round(accuracy_drop, 4)
)
```

---

## **🚨 Detect Concept Drift**

Simple threshold-based detection.

```python
DRIFT_THRESHOLD = 0.10

if accuracy_drop > DRIFT_THRESHOLD:
    print("⚠️ Concept Drift Detected")
else:
    print("✅ No Significant Drift")
```

---

## **📉 Why Concept Drift Happens**

Concept drift occurs when:

- customer behavior changes
- market conditions change
- fraud patterns evolve
- business processes change
- external events affect data

The model was trained on old relationships that no longer hold.

---

## **🧠 Example**

Suppose a loan approval model learned: `Income → Loan Repayment Ability`. During an economic downturn: `Income no longer predicts repayment as strongly`

The relationship changes, causing model performance to drop. This is concept drift.

---

## **🔍 Types of Concept Drift**

### **Sudden Drift**

Performance drops quickly.

Example:

- policy changes
- market crashes

### **Gradual Drift**

Performance slowly decreases over time.

Example:

- changing customer preferences

### **Recurring Drift**

Patterns disappear and later return.

Example:

- seasonal business behaviour

---

## **📊 Production Detection Techniques**

Real-world systems often use:

- rolling accuracy monitoring
- population stability index (PSI)
- statistical hypothesis testing
- drift detection libraries
- shadow model comparisons

This script demonstrates the core idea using performance tracking.

---

## **🧠 Key Takeaways**

1. Concept drift occurs when feature-target relationships change.
2. A model can degrade even if data structure remains the same.
3. Monitoring performance over time helps detect drift.
4. Accuracy decline is often an early warning signal.
5. Detecting drift is a critical ML Systems responsibility.

---

## **Conclusion**

Concept drift is one of the most common reasons machine learning models fail after deployment. Continuous monitoring helps identify when model assumptions are no longer valid, allowing teams to retrain or replace models before performance significantly impacts users.
