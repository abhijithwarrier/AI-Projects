**PYTHON SCRIPT TO **EVALUATE A CLASSIFIER USING CONFUSION MATRIX & ACCURACY SCORE**. üêçü§ñüìä**

This script trains a simple classifier on the Iris dataset and evaluates it using a **confusion matrix** (per-class performance) and **accuracy score** (overall correctness).

### üì¶ Import Libraries
We‚Äôll use scikit-learn for data, model, and metrics; matplotlib to visualize the confusion matrix.

```python
# Import required libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

### üì• Load Dataset & Create Train/Test Split
Use the Iris dataset and keep a hold-out test set to evaluate generalization.

```python
iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

### ü§ñ Train a Simple Classifier
We‚Äôll fit a LogisticRegression model (multiclass One-vs-Rest by default).

```python
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

### üîÆ Make Predictions on the Test Set
Generate predictions to compare against true labels.

```python
y_pred = model.predict(X_test)
```

### üìä Compute Confusion Matrix & Accuracy
Confusion matrix shows per-class hits/misses; accuracy shows overall % correct.

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", round(acc, 4))
print("Confusion Matrix:\n", cm)
```

### üìà Visualize the Confusion Matrix (Optional but nice for IG)
A heatmap makes class-wise performance easy to see at a glance.

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix ‚Äì Logistic Regression on Iris")
plt.tight_layout()
plt.show()
```
