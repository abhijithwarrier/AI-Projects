**PYTHON SCRIPT TO _CREATE AND TRANSFORM FEATURES FOR BETTER TABULAR MACHINE LEARNING PERFORMANCE_. 🧠⚙️📊**

This script demonstrates practical feature engineering techniques used in tabular machine learning:

- creating interaction features
- generating ratio-based features
- applying log transformations
- preparing data for better model learning

---

## **📦 Import Required Libraries**

```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

---

## **🧩 Load Dataset**

```python
data = fetch_california_housing()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

---

## **🚨 Baseline Model**

We first train a model using original features only.

```python
baseline_model = LinearRegression()
baseline_model.fit(X_train, y_train)

baseline_pred = baseline_model.predict(X_test)

print("Baseline R2 Score:", r2_score(y_test, baseline_pred))
```

---

## **🔗 Create Interaction Features**

Interaction features capture relationships between variables.

```python
X_train_fe = X_train.copy()
X_test_fe = X_test.copy()

X_train_fe["MedInc_HouseAge"] = (
    X_train["MedInc"] * X_train["HouseAge"]
)

X_test_fe["MedInc_HouseAge"] = (
    X_test["MedInc"] * X_test["HouseAge"]
)
```

---

## **⚖️ Create Ratio Features**

Ratios often capture useful business or statistical relationships.

```python
X_train_fe["Rooms_Per_Person"] = (
    X_train["AveRooms"] / (X_train["Population"] + 1)
)

X_test_fe["Rooms_Per_Person"] = (
    X_test["AveRooms"] / (X_test["Population"] + 1)
)
```

---

## **📈 Apply Log Transformations**

Log transforms reduce skewness in large-value features.

```python
X_train_fe["Log_Population"] = np.log1p(X_train["Population"])
X_test_fe["Log_Population"] = np.log1p(X_test["Population"])
```

---

## **🤖 Train Model with Engineered Features**

```python
fe_model = LinearRegression()
fe_model.fit(X_train_fe, y_train)

fe_pred = fe_model.predict(X_test_fe)

print("Feature Engineered R2 Score:", r2_score(y_test, fe_pred))
```

---

## **🔍 Why Feature Engineering Matters**

Feature engineering helps models by:

- exposing hidden patterns
- improving signal quality
- reducing noise
- making relationships easier to learn

In many tabular ML tasks, feature engineering matters more than model complexity.

---

## **🧠 Key Takeaways**

1. Feature engineering improves data representation for ML models.
2. Interaction features help capture combined effects.
3. Ratio features reveal hidden relationships.
4. Log transforms help handle skewed distributions.
5. Better features often outperform more complex models.

---

## **Conclusion**

Feature engineering is one of the most important skills in machine learning, especially for tabular data. By transforming and combining features intelligently, we can significantly improve model performance without changing the algorithm itself.
