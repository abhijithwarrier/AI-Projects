**PYTHON SCRIPT TO **_USE THE ELBOW METHOD TO FIND THE OPTIMAL k IN K-MEANS CLUSTERING_**. üêçüìàÔ∏èüìä**

This script demonstrates how to apply the **Elbow Method** to choose the right number of clusters (k) for **K-Means**. We‚Äôll calculate the inertia **(within-cluster sum of squares)** for different values of k and plot them to identify the ‚Äúelbow point.

### üì¶ Import Required Libraries

We‚Äôll use pandas, scikit-learn, numpy, and matplotlib.

```python
# Core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### üßÆ Create a Customer Dataset

We‚Äôll reuse a synthetic dataset with Annual Income and Spending Score.

```python
# Demo dataset
data = {
    "CustomerID": range(1, 21),
    "AnnualIncome_k$": [15, 16, 17, 18, 20, 22, 25, 28, 30, 33, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85],
    "SpendingScore":   [18, 20, 22, 25, 28, 30, 35, 40, 42, 45, 55, 58, 60, 62, 65, 66, 68, 70, 72, 75],
}
df = pd.DataFrame(data)

features = ["AnnualIncome_k$", "SpendingScore"]
X = df[features].values

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### üî¢ Apply K-Means for Different k Values

We loop through values of k and store the inertia.

```python
inertia = []

# Try k from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

### üìà Plot the Elbow Curve

We look for the point where the inertia stops decreasing sharply.

```python
plt.figure(figsize=(7, 5))
plt.plot(range(1, 11), inertia, marker="o", linestyle="--", color="b")
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (WCSS)")
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

### ‚úÖ Notes
- The elbow point (often around k=3‚Äì5 for customer data) indicates a good trade-off.
- Too few clusters ‚Üí underfitting (too broad).
- Too many clusters ‚Üí overfitting (clusters lose meaning).
