**PYTHON SCRIPT TO BUILD A FULL MACHINE LEARNING PIPELINE THAT HANDLES PREPROCESSING AND MODEL TRAINING TOGETHER. 🧠🔧🤖**

This script demonstrates how to build a **production-ready ML pipeline** by combining preprocessing steps and a machine learning model into one clean workflow. This ensures that the same transformations are applied consistently during both training and inference.

---

## **📦 Import Required Libraries**

We’ll use scikit-learn tools for preprocessing, pipeline building, and model training.

```python
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

---

## **🧩 Load the Dataset**

We use a structured tabular dataset suitable for building an end-to-end pipeline.

```python
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **🧪 Introduce Missing Values (for demonstration)**

To simulate a more realistic production scenario, we add a few missing values.

```python
X.iloc[0, 0] = None
X.iloc[10, 3] = None
X.iloc[25, 5] = None
```

---

## **✂️ Split Data into Train and Test Sets**

We keep a separate test set to evaluate the full pipeline properly.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🔧 Build the Full Preprocessing + Model Pipeline**

We combine imputation, scaling, and model training into one workflow.

```python
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", RandomForestClassifier(n_estimators=200, random_state=42))
])
```

This ensures:

- missing values are handled
- numeric features are scaled
- the model is trained using the processed data

---

## **🤖 Train the Pipeline**

```python
pipeline.fit(X_train, y_train)
```

All preprocessing steps are fitted and applied automatically in sequence.

---

## **📊 Evaluate the Pipeline**

We now generate predictions directly from the full pipeline.

```python
y_pred = pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

This gives us end-to-end evaluation using the exact same transformations used during training.

---

## **🚀 Predict on New Data**

The pipeline can also be used on fresh data without manual preprocessing.

```python
sample = X_test.iloc[:5]
predictions = pipeline.predict(sample)

print("Predictions:", predictions)
```

This is exactly why pipelines are essential for production ML systems.

---

## **🔍 Why Full Pipelines Matter**

A full preprocessing + model pipeline helps you:

- avoid inconsistent preprocessing
- prevent data leakage
- simplify training and inference
- make deployment easier
- prepare for hyper-parameter tuning and model persistence

This is one of the most important building blocks in real-world ML engineering.

## **🧠 Key Takeaways**

1. Production ML requires preprocessing and modeling to stay tightly coupled.
2. Pipelines ensure consistent transformations during training and inference.
3. Missing value handling and scaling should be part of the workflow, not separate ad-hoc steps.
4. Pipelines reduce code duplication and deployment mistakes.
5. A full pipeline is a foundational concept for reliable ML systems.
