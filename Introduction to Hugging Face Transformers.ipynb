## **🤖 What is a Pipeline?**

A **pipeline** is the simplest way to use a pretrained AI model.

Instead of manually:

- loading a tokenizer
- loading model weights
- preprocessing input
- generating predictions
- decoding outputs

the pipeline performs all these steps automatically.

---

## **🚀 Create Your First Pipeline**

```python
generator = pipeline(
    task="text-generation",
    model="gpt2"
)
```

This downloads:

- GPT-2 model
- tokenizer
- configuration files

The first run may take a few minutes.

---

## **✍️ Generate Text**

```python
result = generator(
    "Python is one of the most popular",
    max_length=60,
    do_sample=True
)

print(result[0]["generated_text"])
```

The pipeline automatically:

- tokenizes input
- performs inference
- decodes generated tokens
