**PYTHON SCRIPT TO _COMPARE LIGHTGBM AND RANDOMFOREST FOR TABULAR DATA CLASSIFICATION_. 🧠⚔️📊**

This script demonstrates how to compare two powerful ensemble models for structured data:

- **RandomForest** → bagging-based ensemble model
- **LightGBM** → boosting-based high-performance model

We evaluate both models on the same dataset to understand their differences in **accuracy**, **training time**, and overall behaviour.

---

## **📦 Import Required Libraries**

We use pandas for tabular data handling, scikit-learn for dataset splitting and evaluation, and LightGBM / RandomForest for modeling.

```python
import pandas as pd
import time

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

from lightgbm import LGBMClassifier
```

---

## **🧩 Load the Dataset**

We use a structured classification dataset suitable for tabular ML comparison.

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

We keep a separate test set so both models are evaluated fairly on unseen data.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🌲 Train the RandomForest Model**

RandomForest builds many independent trees and combines their predictions.

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

start_rf = time.time()
rf_model.fit(X_train, y_train)
end_rf = time.time()
```

This gives us a strong and stable baseline model.

---

## **🚀 Train the LightGBM Model**

LightGBM builds trees sequentially, with each new tree correcting previous errors.

```python
lgbm_model = LGBMClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42
)

start_lgbm = time.time()
lgbm_model.fit(X_train, y_train)
end_lgbm = time.time()
```

This often leads to better performance and faster training on larger datasets.

---

## **📊 Generate Predictions**

We make both class predictions and probability predictions for each model.

```python
rf_pred = rf_model.predict(X_test)
rf_probs = rf_model.predict_proba(X_test)[:, 1]

lgbm_pred = lgbm_model.predict(X_test)
lgbm_probs = lgbm_model.predict_proba(X_test)[:, 1]
```

This allows us to compare both classification accuracy and probability quality.

---

## **✅ Evaluate Both Models**

We compare the models using accuracy, classification report, ROC-AUC, and training time.

```python
print("=== RandomForest ===")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print("ROC-AUC:", roc_auc_score(y_test, rf_probs))
print("Training Time:", round(end_rf - start_rf, 4), "seconds")
print("\nClassification Report:\n")
print(classification_report(y_test, rf_pred))

print("\n=== LightGBM ===")
print("Accuracy:", accuracy_score(y_test, lgbm_pred))
print("ROC-AUC:", roc_auc_score(y_test, lgbm_probs))
print("Training Time:", round(end_lgbm - start_lgbm, 4), "seconds")
print("\nClassification Report:\n")
print(classification_report(y_test, lgbm_pred))
```

This gives a direct practical comparison of performance and efficiency.

---

## **🔍 Why This Comparison Matters**

Both models are excellent for tabular data, but they differ in approach:

### **🌲 RandomForest**

- uses **bagging**
- trains trees independently
- strong baseline
- stable and simple to tune

### **🚀 LightGBM**

- uses **boosting**
- trains trees sequentially
- often faster and more accurate
- better suited for high-performance tabular ML

Understanding these trade-offs helps you choose the right model for the right problem.

---

## **🧠 Key Takeaways**

1. RandomForest and LightGBM are both strong models for tabular data.
2. RandomForest uses bagging, while LightGBM uses boosting.
3. LightGBM often trains faster and achieves stronger performance.
4. RandomForest is simpler and more stable as a baseline.
5. Choosing between them depends on speed, scale, and performance needs.

Comparing LightGBM and RandomForest is a practical way to understand how different ensemble strategies behave on tabular datasets. While RandomForest provides simplicity and stability, LightGBM offers speed and higher modeling power for advanced use cases.
