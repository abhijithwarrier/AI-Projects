**PYTHON SCRIPT TO _IMPLEMENT A SIMPLE MODEL VERSIONING AND ROLLBACK STRATEGY FOR PRODUCTION MACHINE LEARNING SYSTEMS_. 🧠📦🔁**

This script demonstrates how to manage multiple model versions in a production-style workflow. We simulate:

- **v1** → currently deployed model
- **v2** → newer candidate model
- **rollback** → reverting to v1 if v2 underperforms

Model versioning is essential for safe deployment, monitoring, and recovery in real-world ML systems.

---

## **📦 Import Required Libraries**

We use scikit-learn for training models and joblib for saving different model versions.

```python
import joblib
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

---

## **🧩 Load the Dataset**

We use a structured classification dataset to simulate training multiple model versions.

```python
data = load_wine()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

We keep a separate test set to evaluate each model version fairly.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🤖 Train Model Version 1 (v1)**

This represents the currently deployed model.

```python
model_v1 = RandomForestClassifier(
    n_estimators=100,
    max_depth=4,
    random_state=42
)

model_v1.fit(X_train, y_train)
```

---

## **💾 Save Version 1**

We save the model with an explicit versioned filename.

```python
joblib.dump(model_v1, "model_v1.pkl")
```

This makes it easier to track which version is deployed.

---

## **🚀 Train Model Version 2 (v2)**

This represents a new candidate model with updated hyperparameters.

```python
model_v2 = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

model_v2.fit(X_train, y_train)
```

---

## **💾 Save Version 2**

```python
joblib.dump(model_v2, "model_v2.pkl")
```

Now we have two model versions available for comparison.

---

## **📊 Evaluate Both Versions**

We compare performance on the same test data.

```python
pred_v1 = model_v1.predict(X_test)
pred_v2 = model_v2.predict(X_test)

acc_v1 = accuracy_score(y_test, pred_v1)
acc_v2 = accuracy_score(y_test, pred_v2)

print("v1 Accuracy:", acc_v1)
print("v2 Accuracy:", acc_v2)
```

This helps decide whether the newer model should replace the current one.

---

## **✅ Select Active Model**

If v2 performs better, promote it. Otherwise, keep v1.

```python
if acc_v2 >= acc_v1:
    active_model = model_v2
    active_version = "v2"
else:
    active_model = model_v1
    active_version = "v1"

print("Active model version:", active_version)
```

This simulates a simple deployment decision rule.

---

## **🔁 Simulate Rollback Strategy**

If the new model fails later in production, we can reload the older version.

```python
rollback_model = joblib.load("model_v1.pkl")
rollback_preds = rollback_model.predict(X_test)

print("Rollback model accuracy:", accuracy_score(y_test, rollback_preds))
```

This shows how rollback restores a known-good version quickly.

---

## **🗂️ Why Versioning Matters**

A versioning strategy helps you:

- track deployed model history
- compare old vs new versions
- roll back safely when issues occur
- support reproducibility and auditing
- reduce deployment risk

Without versioning, production ML systems become fragile and difficult to manage.

---

## **🧠 Key Takeaways**

1. Production ML systems should never rely on a single unnamed model file.
2. Versioned models make comparison and deployment safer.
3. Evaluating v1 vs v2 helps decide promotion strategy.
4. Rollback is essential when a new version underperforms.
5. Model versioning is a foundational production ML practice.
