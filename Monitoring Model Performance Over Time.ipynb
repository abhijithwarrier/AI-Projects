**PYTHON SCRIPT TO TRACK MACHINE LEARNING MODEL PERFORMANCE ACROSS MULTIPLE TIME PERIODS. 🧠📈⏱️**

This script demonstrates how to monitor model performance over time using simulated daily metrics such as:

- accuracy
- precision
- recall
- F1-score
- average confidence

This helps identify performance degradation, drift, and when retraining may be needed.

---

## **📦 Import Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

---

## **📝 Create Performance Tracking Data**

We simulate model performance metrics collected over several days.

```python
performance_data = pd.DataFrame({
    "date": pd.date_range(
        start="2025-01-01",
        periods=10,
        freq="D"
    ),

    "accuracy": [
        0.96, 0.96, 0.95, 0.94, 0.94,
        0.92, 0.91, 0.90, 0.89, 0.87
    ],

    "precision": [
        0.95, 0.95, 0.94, 0.93, 0.92,
        0.91, 0.90, 0.89, 0.87, 0.85
    ],

    "recall": [
        0.94, 0.93, 0.93, 0.92, 0.91,
        0.90, 0.88, 0.87, 0.86, 0.84
    ],

    "f1_score": [
        0.945, 0.94, 0.935, 0.925, 0.915,
        0.905, 0.89, 0.88, 0.865, 0.845
    ],

    "avg_confidence": [
        0.94, 0.94, 0.93, 0.92, 0.91,
        0.90, 0.89, 0.88, 0.86, 0.84
    ]
})
```

---

## **🔍 View Monitoring Data**

```python
print("Model Performance Tracking Data:")
print(performance_data)
```

---

## **📈 Plot Accuracy Over Time**

```python
plt.figure(figsize=(8, 4))

plt.plot(
    performance_data["date"],
    performance_data["accuracy"],
    marker="o"
)

plt.title("Model Accuracy Over Time")
plt.xlabel("Date")
plt.ylabel("Accuracy")
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

## **📈 Plot Precision, Recall, and F1-score**

```python
plt.figure(figsize=(8, 4))

plt.plot(
    performance_data["date"],
    performance_data["precision"],
    marker="o",
    label="Precision"
)

plt.plot(
    performance_data["date"],
    performance_data["recall"],
    marker="o",
    label="Recall"
)

plt.plot(
    performance_data["date"],
    performance_data["f1_score"],
    marker="o",
    label="F1 Score"
)

plt.title("Classification Metrics Over Time")
plt.xlabel("Date")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

## **📉 Plot Average Confidence Over Time**

```python
plt.figure(figsize=(8, 4))

plt.plot(
    performance_data["date"],
    performance_data["avg_confidence"],
    marker="o"
)

plt.title("Average Prediction Confidence Over Time")
plt.xlabel("Date")
plt.ylabel("Average Confidence")
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

## **🚨 Detect Performance Drop**

We add simple alert logic when performance drops below a threshold.

```python
latest_accuracy = performance_data["accuracy"].iloc[-1]
latest_f1 = performance_data["f1_score"].iloc[-1]

if latest_accuracy < 0.90:
    print("⚠️ Alert: Accuracy dropped below threshold")

if latest_f1 < 0.85:
    print("⚠️ Alert: F1-score dropped below threshold")
```

---

## **📊 Calculate Performance Change**

```python
accuracy_change = (
    performance_data["accuracy"].iloc[-1]
    - performance_data["accuracy"].iloc[0]
)

f1_change = (
    performance_data["f1_score"].iloc[-1]
    - performance_data["f1_score"].iloc[0]
)

print("Accuracy Change:", round(accuracy_change, 4))
print("F1-score Change:", round(f1_change, 4))
```

## **🔍 Why Performance Tracking Matters**

Monitoring model performance over time helps detect:

- performance degradation
- feature drift
- concept drift
- data quality issues
- need for retraining

This is a key part of ML system reliability.

---

## **🧠 Key Takeaways**

1. Model performance should be monitored continuously after deployment.
2. Metrics like accuracy, precision, recall, and F1-score reveal model health.
3. Confidence trends help detect uncertainty changes.
4. Alerts help catch degradation early.
5. Performance tracking is a core MLOps practice.

---

## **Conclusion**

Monitoring model performance over time is essential for production ML systems. A model that performs well today may degrade as data and user behavior change. Tracking metrics over time helps detect issues early and supports better retraining decisions.
