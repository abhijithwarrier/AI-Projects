Named Entity Recognition (NER) is one of the most powerful Natural Language Processing (NLP) tasks. Instead of simply understanding text, AI can identify important entities such as **people, organisations, locations, dates, and products** within a sentence.

## 📦 **Import Required Libraries**

````python
from transformers import pipeline
````

---

## 🤖 **Create NER Pipeline**

````python
ner = pipeline(
    task="ner",
    grouped_entities=True
)
````

The pipeline automatically downloads a pre-trained Named Entity Recognition model.

---

## 📝 **Create Sample Text**

````python
text = "Sundar Pichai is the CEO of Google and spoke at an AI conference in California on July 15, 2025."
````

---

## 🚀 **Extract Named Entities**

````python
entities = ner(text)
````

---

## 📊 **Display Results**

````python
print("Input Text:\n")
print(text)
print("\nDetected Entities:\n")

for entity in entities:
    print(f"Entity: {entity['word']}")
    print(f"Type: {entity['entity_group']}")
    print(f"Confidence: {entity['score']:.4f}")
    print("-" * 40)
````

---

## 📄 **Process Multiple Sentences**

````python
documents = [
    "Microsoft announced new AI services in Seattle.",
    "Elon Musk founded SpaceX.",
    "The Olympics will be held in Los Angeles.",
    "Amazon opened a new office in London."
]
````

---

## 🔄 **Run NER on Multiple Documents**

````python
for doc in documents:
    print("=" * 60)
    print(doc)

    results = ner(doc)
    for entity in results:
        print(entity["word"], "→", entity["entity_group"])
````
