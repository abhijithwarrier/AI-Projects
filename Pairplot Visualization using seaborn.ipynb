This script demonstrates how to use **Seabornâ€™s Pairplot** to visualize pairwise relationships between numerical features in a dataset.

Pairplots help uncover patterns, clusters, correlations, and class separations â€” making them a powerful tool for exploratory data analysis (EDA).

---

## **ðŸ“¦ Import Required Libraries**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

---

## **ðŸ§© Load a Sample Dataset**

Weâ€™ll use the **Iris dataset**, a classic multi-class dataset with four numerical features.

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
df.head()
```

To make the plot readable, we convert numeric labels â†’ text names:

```python
df["species"] = df["species"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
```

---

## **ðŸ“Š Create a Pairplot with Seaborn**

A pairplot draws scatter plots for each pair of numerical features and histograms (or KDEs) on the diagonal.

```python
sns.pairplot(
    df,
    hue="species",
    diag_kind="kde",
    corner=True,        # cleaner grid (optional)
    palette="viridis"   # color theme
)
plt.suptitle("Iris Dataset â€“ Pairplot Visualization", y=1.02)
plt.show()
```

---

## **ðŸŽ¯ What You Can Observe**

- **Cluster separation** between species.
- **Correlated features** (e.g., petal length vs petal width).
- **Feature distributions** of each species.
- **Outliers & trends** across numeric dimensions.

Pairplots are an essential first step before training ML models â€” they help you understand what features matter.
