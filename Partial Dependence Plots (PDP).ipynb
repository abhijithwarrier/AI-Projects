**PYTHON SCRIPT TO _UNDERSTAND HOW FEATURES AFFECT MODEL PREDICTIONS USING PARTIAL DEPENDENCE PLOTS_. üß†üìàüîç**

This script demonstrates how **Partial Dependence Plots (PDP)** help visualise the **average effect of one or more features** on a machine learning model‚Äôs predictions, while marginalizing over all other features.

---

## **üì¶ Import Required Libraries**

We‚Äôll use scikit-learn‚Äôs inspection tools along with a tree-based model.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import PartialDependenceDisplay
```

---

## **üß© Load the Dataset**

We use a tabular dataset suitable for interpretability analysis.

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **‚úÇÔ∏è Split Data into Train and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **ü§ñ Train a Machine Learning Model**

We train a Random Forest classifier for demonstration.

```python
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)
model.fit(X_train, y_train)
```

---

## **üìà Generate Partial Dependence Plots**

We visualise how individual features influence the model‚Äôs predictions.

```python
features = ["mean radius", "mean texture"]

PartialDependenceDisplay.from_estimator(
    model,
    X_train,
    features=features,
    kind="average"
)

plt.tight_layout()
plt.show()
```

Each curve shows the **average predicted response** as the feature value changes.

---

## **üîç Why Partial Dependence Plots Are Useful**

- Show **directional influence** of a feature
- Help identify **non-linear effects**
- Useful for **model debugging and trust**
- Complement SHAP and permutation importance

PDPs answer the question:

*‚ÄúOn average, what happens to predictions when this feature changes?‚Äù*

---

## **üß† Key Takeaways**

1. Partial Dependence Plots visualize feature impact on predictions.
2. They marginalize the effect of all other features.
3. PDPs help detect non-linear and threshold effects.
4. Best used with low-correlated features.
5. A powerful interpretability tool alongside SHAP.

---

## **Conclusion**

Partial Dependence Plots provide an intuitive way to understand how individual features influence machine learning models. By visualizing average prediction trends, PDPs make complex models more transparent and easier to trust.
