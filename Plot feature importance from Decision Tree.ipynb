**PYTHON SCRIPT TO DEMONSTRATE **PLOTTING FEATURE IMPORTANCE FROM DECISION TREE CLASSIFIER**. üêçüìäü§ñ**

This script demonstrates how to extract and visualize **feature importance** scores from a **trained Decision Tree model**, helping us understand which features contribute most to **predictions**.

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

### üìä Load Dataset & Train Model
We'll use the same Iris dataset and train a Decision Tree Classifier.

```python
# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names

# Train Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)
```

### üîç Extract Feature Importance
The Decision Tree provides a `feature_importances_` attribute that
gives importance scores for each feature.

```python
# Extract feature importance
importance = model.feature_importances_
df_importance = pd.DataFrame({'Feature': feature_names, 'Importance': importance})
df_importance = df_importance.sort_values(by='Importance', ascending=False)
df_importance
```

### üìà Plot Feature Importance
We'll plot a horizontal bar chart for better readability.

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.barh(df_importance['Feature'], df_importance['Importance'], color='teal')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importance from Decision Tree')
plt.gca().invert_yaxis()
plt.show()
```
