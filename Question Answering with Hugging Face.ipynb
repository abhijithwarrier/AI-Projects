## **📦 Import Required Libraries**

```python
from transformers import pipeline
```

---

## **🤖 Create a Question Answering Pipeline**

```python
qa_pipeline = pipeline(
    task="question-answering"
)
```

The pipeline automatically downloads:

- pretrained QA model
- tokenizer
- configuration

---

## **📖 Create Context**

```python
context = """
Artificial Intelligence (AI) is the simulation of human intelligence by machines.
Machine Learning is a subset of AI that allows systems to learn from data.
Deep Learning is a subset of Machine Learning that uses neural networks with multiple layers.
Hugging Face provides pretrained transformer models for NLP tasks.
"""
```

---

## **❓ Ask a Question**

```python
question = (
    "What is Machine Learning?"
)
```

---

## **🚀 Generate Answer**

```python
result = qa_pipeline(
    question=question,
    context=context
)
```

---

## **📊 Display Result**

```python
print("Question:")
print(question)

print("\nAnswer:")
print(result["answer"])

print("\nConfidence:")
print(round(result["score"], 4))
```

---

## **🔄 Try Multiple Questions**

```python
questions = [
    "What is AI?",
    "What is Deep Learning?",
    "What does Hugging Face provide?"
]

for q in questions:
    result = qa_pipeline(
        question=q,
        context=context
    )

    print("=" * 50)
    print("Question:", q)
    print("Answer:", result["answer"])
    print("Confidence:", round(result["score"], 4))
```

---

## **🧠 How Question Answering Works**

Unlike text generation, Question Answering models:

1. Read the supplied context.
2. Understand the user’s question.
3. Identify the relevant span of text.
4. Return the most likely answer.

The answer is extracted **from the context itself**, not generated from scratch.
