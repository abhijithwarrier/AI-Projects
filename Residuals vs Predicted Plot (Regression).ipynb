**PYTHON SCRIPT TO _ANALYSE REGRESSION MODEL ERRORS USING RESIDUALS VS PREDICTED PLOTS_. 🧠📊📉**

This script demonstrates how to evaluate regression model performance by visualising **residuals (prediction errors)** against predicted values. Residual plots help detect patterns that indicate model issues such as non-linearity, heteroscedasticity, or outliers.

---

## **📦 Import Required Libraries**

We’ll use scikit-learn utilities for regression and matplotlib for visualisation.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

---

## **🧩 Load the Dataset**

We use a regression dataset suitable for error analysis.

```python
data = fetch_california_housing()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

---

## **🤖 Train the Regression Model**

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

---

## **📊 Generate Predictions**

```python
y_pred = model.predict(X_test)
```

---

## **📉 Compute Residuals**

Residuals represent the difference between actual and predicted values.

```python
residuals = y_test - y_pred
```

---

## **📈 Plot Residuals vs Predicted Values**

```python
plt.figure(figsize=(6, 6))

plt.scatter(y_pred, residuals, alpha=0.6)

plt.axhline(0, linestyle="--")

plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted Plot")

plt.show()
```

---

## **🔍 How to Interpret the Plot**

- Random scatter around zero → good model fit
- Curved pattern → missing non-linear relationship
- Funnel shape → heteroscedasticity (variance changes)
- Extreme points → possible outliers

Residual plots are essential diagnostics for regression models.

---

## **🧠 Key Takeaways**

1. Residuals measure prediction error.
2. Random scatter indicates a well-fitted model.
3. Patterns reveal model misspecification.
4. Funnel shapes indicate variance issues.
5. A key diagnostic tool for regression analysis.

---

## **Conclusion**

Residuals vs predicted plots provide a simple yet powerful way to diagnose regression model performance. By visualising prediction errors, we can detect patterns that indicate underlying issues and improve model reliability.
