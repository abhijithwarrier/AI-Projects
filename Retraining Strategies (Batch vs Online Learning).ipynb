**PYTHON SCRIPT TO _COMPARE BATCH RETRAINING AND ONLINE LEARNING STRATEGIES IN MACHINE LEARNING SYSTEMS_. 🧠🔄📈**

This script demonstrates two common approaches used to keep machine learning models updated after deployment:

- **Batch Retraining** → periodically retrain using accumulated data
- **Online Learning** → continuously update the model as new data arrives

Understanding when to use each strategy is critical for production ML systems.

---

## **📦 Import Required Libraries**

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
```

---

## **🧩 Load Dataset**

```python
data = load_breast_cancer()

X = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

y = data.target
```

---

## **✂️ Split Initial and Future Data**

We simulate future incoming data.

```python
X_initial, X_future, y_initial, y_future = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

scaler = StandardScaler()

X_initial = scaler.fit_transform(X_initial)
X_future = scaler.transform(X_future)
```

---

# **🧠 Part 1 – Batch Retraining**

## **🤖 Train Initial Model**

```python
batch_model = SGDClassifier(
    random_state=42
)

batch_model.fit(
    X_initial,
    y_initial
)
```

---

## **📊 Evaluate Initial Performance**

```python
initial_pred = batch_model.predict(
    X_future
)

print(
    "Initial Accuracy:",
    accuracy_score(
        y_future,
        initial_pred
    )
)
```

---

## **🔄 Simulate Batch Retraining**

After collecting new data, retrain the model.

```python
batch_model.fit(
    X,
    y
)
```

This replaces the previous model with a newly trained version.

---

## **📊 Evaluate Retrained Model**

```python
retrained_pred = batch_model.predict(
    X_future
)

print(
    "Batch Retrained Accuracy:",
    accuracy_score(
        y_future,
        retrained_pred
    )
)
```

---

# **🧠 Part 2 – Online Learning**

Online learning updates the model incrementally.

---

## **🤖 Create Online Model**

```python
online_model = SGDClassifier(
    random_state=42,
    loss="log_loss"
)
```

---

## **🚀 Initial Training**

```python
online_model.partial_fit(
    X_initial,
    y_initial,
    classes=[0, 1]
)
```

---

## **🔄 Simulate Streaming Updates**

Process incoming batches one at a time.

```python
batch_size = 20

for start in range(
    0,
    len(X_future),
    batch_size
):

    end = start + batch_size

    X_batch = X_future[start:end]
    y_batch = y_future[start:end]

    online_model.partial_fit(
        X_batch,
        y_batch
    )
```

The model continuously learns from new data.

---

## **📊 Evaluate Online Model**

```python
online_pred = online_model.predict(
    X_future
)

print(
    "Online Learning Accuracy:",
    accuracy_score(
        y_future,
        online_pred
    )
)
```

---

## **⚖️ Batch vs Online Learning**

### **🔹 Batch Retraining**

- retrains on entire dataset
- often higher accuracy
- computationally expensive
- periodic updates

Best for:

- nightly retraining
- weekly retraining
- stable environments

---

### **🔹 Online Learning**

- updates incrementally
- adapts quickly
- lower training cost
- supports streaming data

Best for:

- real-time systems
- recommendation engines
- fraud detection
- dynamic environments

---

## **🔍 Why Retraining Matters**

Models degrade because of:

- concept drift
- changing user behavior
- new business conditions
- evolving data distributions

Retraining keeps models relevant.

---

## **🧠 Key Takeaways**

1. Retraining helps combat model degradation.
2. Batch retraining rebuilds models periodically.
3. Online learning updates models continuously.
4. Different applications require different strategies.
5. Retraining is a critical part of production ML systems.

---

## **Conclusion**

Machine learning models are not static assets. As data changes, models must evolve as well. Choosing the right retraining strategy—batch or online—depends on business requirements, infrastructure constraints, and how quickly data patterns change.
