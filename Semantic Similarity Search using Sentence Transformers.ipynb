Traditional keyword search relies on matching exact words, which often fails when different words express the same idea. **Semantic Similarity Search** solves this by comparing the *meaning* of text instead of just matching keywords.

In this project, we'll build a **Semantic Similarity Search** application using **Sentence Transformers**.

We'll learn how to:

- Generate sentence embeddings
- Compute semantic similarity
- Find the most relevant documents
- Rank search results
- Build the foundation for semantic search and RAG systems

## 📦 **Import Required Libraries**

````python
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
````

---

## 🤖 **Load a Pretrained Embedding Model**

````python
model = SentenceTransformer("all-MiniLM-L6-v2")
````

This lightweight model converts text into dense numerical vectors (embeddings) that capture semantic meaning.

---

## 📄 **Create Sample Documents**

````python
documents = [
    "Python is widely used for machine learning and AI.",
    "Deep learning enables computers to recognize images.",
    "Transformers power modern large language models.",
    "Pandas is a powerful library for data analysis.",
    "Neural networks are inspired by the human brain.",
    "Football is one of the most popular sports."
]
````

---

## 🔢 **Generate Document Embeddings**

````python
document_embeddings = model.encode(
    documents,
    convert_to_tensor=True
)
````

---

## 🔍 **Create a Search Query**

````python
query = "How are AI models built?"
````

---

## 🚀 **Generate Query Embedding**

````python
query_embedding = model.encode(query, convert_to_tensor=True)
````

---

## 📊 **Compute Semantic Similarities**

````python
similarities = util.cos_sim(query_embedding, document_embeddings)
````

---

## 📈 **Rank Documents**

````python
scores = similarities[0]

ranked_results = sorted(
    zip(documents, scores),
    key=lambda x: x[1],
    reverse=True
)
````

---

## 📋 **Display Search Results**

````python
print("Search Query:\n")
print(query)
print("\nMost Relevant Documents:\n")

for document, score in ranked_results:
    print("=" * 60)
    print(document)
    print(f"\nSimilarity Score: {score:.4f}")
````

---

## 🔄 **Try Multiple Queries**

````python
queries = [
    "Artificial Intelligence",
    "Data Science",
    "Sports",
    "Neural Networks"
]

for query in queries:
    print("\n" + "#" * 70)
    print("Query:", query)

    query_embedding = model.encode(query, convert_to_tensor=True)
    similarities = util.cos_sim(query_embedding, document_embeddings)[0]
    best_match = similarities.argmax()

    print("Best Match:", documents[best_match])
````
