Modern AI applications don't simply search for matching keywords—they understand the **meaning** behind text. This capability is made possible through **Sentence Embeddings**, where entire sentences are converted into dense numerical vectors that capture semantic information.

Sentence Transformers provide one of the easiest and most powerful ways to generate these embeddings. They form the backbone of semantic search engines, Retrieval-Augmented Generation (RAG), recommendation systems, duplicate detection, document clustering, enterprise search, and many other Large Language Model applications.

In this project, we'll build a simple Sentence Embeddings application using the **Sentence Transformers** library.

We'll learn how to:

- Generate sentence embeddings
- Compare semantic similarity
- Find the most relevant sentences
- Perform semantic search
- Understand how embeddings power modern LLM applications

## 📦 Import Required Libraries

````python
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
````

The **SentenceTransformer** class loads pretrained embedding models, while **util** provides helper functions such as cosine similarity for comparing embeddings.

---

## 🤖 Load a Pretrained Sentence Transformer

````python
model = SentenceTransformer("all-MiniLM-L6-v2")
````

The **all-MiniLM-L6-v2** model is a lightweight yet highly accurate pretrained model capable of generating meaningful sentence embeddings suitable for most semantic AI applications.

---

## 📝 Create Sample Sentences

````python
sentences = [
    "Python is widely used for Artificial Intelligence.",
    "Machine learning allows computers to learn from data.",
    "Large Language Models understand natural language.",
    "Football is played by millions of people worldwide.",
    "Deep learning uses neural networks.",
    "Pandas is a powerful Python library for data analysis."
]
````

These sentences will act as our small knowledge base that we'll later search using semantic similarity instead of keyword matching.

---

## 🔢 Generate Sentence Embeddings

````python
embeddings = model.encode(
    sentences,
    convert_to_tensor=True
)
````

Every sentence is converted into a high-dimensional numerical vector. These vectors preserve the semantic meaning of the original text, allowing similar ideas to be placed close together in vector space.

---

## 🔍 Encode a Search Query

````python
query = "Artificial Intelligence with Python"

query_embedding = model.encode(
    query,
    convert_to_tensor=True
)
````

Just like the stored sentences, our search query must also be converted into an embedding before any similarity comparison can be performed.

---

## 📊 Compute Semantic Similarity

````python
similarities = util.cos_sim(
    query_embedding,
    embeddings
)
````

Cosine Similarity measures how closely two sentence embeddings are related. Scores closer to **1** indicate stronger semantic similarity, while scores closer to **0** indicate weaker relationships.

---

## 📈 Rank the Most Similar Sentences

````python
scores = similarities[0]

ranked_results = sorted(
    zip(sentences, scores),
    key=lambda x: x[1],
    reverse=True
)
````

The similarity scores are paired with their corresponding sentences and sorted in descending order so that the most relevant matches appear first.

---

## 📋 Display Search Results

````python
print("Search Query:\n")
print(query)

print("\nMost Similar Sentences:\n")

for sentence, score in ranked_results:
    print("=" * 60)
    print(sentence)
    print(f"\nSimilarity Score: {score:.4f}")
````

Running this code displays the sentences ordered by semantic similarity, allowing us to retrieve information based on meaning rather than exact wording.

---

## 🔄 Compare Multiple Search Queries

````python
queries = [
    "Neural Networks",
    "Artificial Intelligence",
    "Sports",
    "Python Programming"
]

for query in queries:
    print("\n" + "#" * 70)
    print("Query:", query)

    query_embedding = model.encode(
        query,
        convert_to_tensor=True
    )

    similarities = util.cos_sim(
        query_embedding,
        embeddings
    )[0]

    best_match = similarities.argmax()

    print("Best Match:", sentences[best_match])
````

Instead of searching with only one query, we can repeatedly encode different search phrases and retrieve the closest matching sentence each time. This forms the basis of semantic search engines and intelligent knowledge retrieval systems.

---

## 📊 Compare Two Individual Sentences

````python
sentence1 = "Python is excellent for AI."
sentence2 = "Artificial Intelligence projects commonly use Python."

embedding1 = model.encode(
    sentence1,
    convert_to_tensor=True
)

embedding2 = model.encode(
    sentence2,
    convert_to_tensor=True
)

score = util.cos_sim(
    embedding1,
    embedding2
)

print("\nSimilarity Between Two Sentences")
print("=" * 60)
print(sentence1)
print(sentence2)
print(f"\nSimilarity Score: {score.item():.4f}")
````

Even though the wording differs, these two sentences express nearly the same idea. Sentence embeddings capture this semantic relationship far better than traditional keyword matching techniques.
