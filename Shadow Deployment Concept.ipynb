**PYTHON SCRIPT TO SIMULATE SHADOW DEPLOYMENT BY RUNNING A CANDIDATE MODEL ALONGSIDE A PRODUCTION MODEL. 🧠🌗🚀**

This script demonstrates the concept of **Shadow Deployment**, where a new model receives the same input as the production model but does not affect the final user-facing decision.

The candidate model runs silently in the background so teams can compare its behaviour safely before rollout.

---

## **📦 Import Required Libraries**

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, f1_score
```

---

## **🧩 Load Dataset**

```python
data = load_breast_cancer()

X = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

y = data.target
```

---

## **✂️ Split Dataset**

```python
X_train, X_live, y_train, y_live = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42,
    stratify=y
)
```

Here, `X_live` simulates live production requests.

---

## **🤖 Train Production Model**

```python
production_model = LogisticRegression(
    max_iter=5000
)

production_model.fit(
    X_train,
    y_train
)
```

This represents the current deployed model.

---

## **🌗 Train Shadow Model**

```python
shadow_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

shadow_model.fit(
    X_train,
    y_train
)
```

This represents the candidate model running silently in the background.

---

## **🚀 Simulate Live Requests**

```python
live_requests = X_live.iloc[:10]
live_labels = y_live[:10]
```

Both models will receive the same live inputs.

---

## **📊 Generate Production Predictions**

```python
production_predictions = production_model.predict(
    live_requests
)
```

These are the predictions that would be used in production.

---

## **🌗 Generate Shadow Predictions**

```python
shadow_predictions = shadow_model.predict(
    live_requests
)
```

These predictions are logged only and do not affect users.

---

## **📝 Create Shadow Deployment Logs**

```python
shadow_logs = pd.DataFrame({
    "actual": live_labels,
    "production_prediction": production_predictions,
    "shadow_prediction": shadow_predictions
})
```

---

## **🔍 Compare Model Outputs**

```python
shadow_logs["prediction_match"] = (
    shadow_logs["production_prediction"]
    ==
    shadow_logs["shadow_prediction"]
)

print(shadow_logs)
```

This helps identify where the shadow model behaves differently.

---

## **📈 Evaluate Both Models**

```python
prod_accuracy = accuracy_score(
    live_labels,
    production_predictions
)

shadow_accuracy = accuracy_score(
    live_labels,
    shadow_predictions
)

prod_f1 = f1_score(
    live_labels,
    production_predictions
)

shadow_f1 = f1_score(
    live_labels,
    shadow_predictions
)

print("Production Accuracy:", prod_accuracy)
print("Shadow Accuracy:", shadow_accuracy)

print("Production F1:", prod_f1)
print("Shadow F1:", shadow_f1)
```

---

## **🔍 Why Shadow Deployment Matters**

Shadow deployment helps teams:

- test new models safely
- compare production vs candidate behaviour
- detect unexpected prediction differences
- collect real-world performance signals
- reduce deployment risk

The user-facing system continues using only the production model.

---

## **🧠 Key Takeaways**

1. Shadow deployment runs a candidate model beside the production model.
2. Shadow predictions are logged but not served to users.
3. It helps compare behaviour safely on live-like data.
4. Useful before A/B testing or full rollout.
5. A key deployment strategy in production ML systems.

---

## **Conclusion**

Shadow deployment is a safe way to validate new machine learning models before exposing them to users. By running the candidate model silently in parallel with the production model, teams can observe behaviour, compare predictions, and reduce rollout risk.
