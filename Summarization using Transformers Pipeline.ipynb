## **📦 Import Required Libraries**

````python
from transformers import pipeline
````

---

## **🤖 Create Summarization Pipeline**

````python
summarizer = pipeline(
    task="summarization"
)
````

The first execution downloads a pretrained summarization model.

---

## **📄 Create Sample Document**

````python
document = """
Artificial Intelligence has transformed many industries over the past decade.
Machine Learning allows computers to learn from data without explicit programming.
Deep Learning has enabled significant advances in computer vision and natural language processing.
Today, transformer models power applications such as chatbots, document summarization,
translation, recommendation systems, and intelligent search engines. As organizations
adopt AI at scale, the demand for efficient language models and automation continues to grow.
"""
````

---

## **🚀 Generate Summary**

````python
summary = summarizer(
    document,
    max_length=60,
    min_length=20,
    do_sample=False
)
````

---

## **📖 Display Results**

````python
print("Original Document:\n")
print(document)

print("\nSummary:\n")
print(summary[0]["summary_text"])
````

---

## **🎛️ Understanding Summarization Parameters**

### **max_length**

Controls the maximum length of the summary.

````python
max_length=60
````

### **min_length**

Prevents summaries from becoming too short.

````python
min_length=20
````

### **do_sample**

````python
do_sample=False
````

Using `False` makes the output more deterministic and suitable for factual summaries.

---

## **🔄 Try Another Document**

````python
another_document = """
Python has become one of the world's most popular programming languages due to its
simple syntax, extensive ecosystem, and versatility. It is widely used in web
development, automation, artificial intelligence, machine learning, scientific
computing, and data analysis.
"""

summary = summarizer(
    another_document,
    max_length=40,
    min_length=15,
    do_sample=False
)

print(summary[0]["summary_text"])
````
