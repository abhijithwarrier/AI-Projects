Text Classification is one of the most common Natural Language Processing (NLP) tasks. It involves assigning predefined categories to text, such as determining sentiment, identifying spam, categorizing news articles, or detecting the topic of a document.

In this project, we’ll build a **Text Classification** application using the Hugging Face Transformers pipeline.

We’ll learn how to:

- Load a pretrained text classification model
- Classify text into categories
- Interpret confidence scores
- Process multiple documents
- Explore real-world NLP classification tasks

## **📦 Import Required Libraries**

````python
from transformers import pipeline
````

---

## **🤖 Create Text Classification Pipeline**

````python
classifier = pipeline(task="text-classification")
````

The pipeline automatically downloads a pretrained classification model during its first execution.

---

## **📝 Classify a Single Sentence**

````python
text = "Python is an amazing programming language for AI development."

result = classifier(text)
````

---

## **📊 Display Prediction**

````python
print("Input Text:\n")
print(text)

print("\nPrediction:\n")
print(result)
````

Typical output:

````python
[
    {
        'label': 'POSITIVE',
        'score': 0.9998
    }
]
````

---

## **📄 Classify Multiple Documents**

````python
documents = [
    "The customer service was excellent.",
    "The delivery was delayed and disappointing.",
    "The product quality exceeded my expectations.",
    "I will never buy this product again."
]

results = classifier(documents)
````

---

## **📋 Display Results**

````python
for text, prediction in zip(
    documents,
    results
):
    print("=" * 60)

    print("Text:")
    print(text)

    print("\nLabel:")
    print(prediction["label"])

    print("Confidence:")
    print(round(prediction["score"], 4))
````

---

## **📈 Understanding the Output**

Each prediction contains:

- **label** → Predicted category
- **score** → Confidence score

Example:

```jsx
{
    "label": "POSITIVE",
    "score": 0.9987
}
```

A higher confidence score indicates the model is more certain about its prediction.
