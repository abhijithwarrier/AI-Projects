## **📦 Import Required Libraries**

```python
from transformers import (
    GPT2Tokenizer,
    GPT2LMHeadModel
)

import torch
```

---

## **🤖 Load GPT-2 Tokenizer**

```python
tokenizer = GPT2Tokenizer.from_pretrained(
    "gpt2"
)
```

The tokenizer converts human-readable text into tokens that GPT-2 understands.

---

## **🧠 Load GPT-2 Model**

```python
model = GPT2LMHeadModel.from_pretrained(
    "gpt2"
)
```

This downloads the pretrained GPT-2 language model.

The first run may take a few minutes.

---

## **✍️ Create a Prompt**

```python
prompt = (
    "Artificial Intelligence is transforming"
)
```

This serves as the starting point for text generation.

---

## **🔤 Tokenize the Prompt**

```python
inputs = tokenizer.encode(
    prompt,
    return_tensors="pt"
)
```

The text is converted into numerical tokens.

---

## **🚀 Generate Text**

```python
outputs = model.generate(

    inputs,

    max_length=80,

    temperature=0.8,

    do_sample=True,

    top_k=50,

    top_p=0.95
)
```

The model predicts one token at a time until the desired length is reached.

---

## **📖 Decode Generated Tokens**

```python
generated_text = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True
)

print(generated_text)
```

The generated tokens are converted back into readable text.

---

## **🎛️ Understanding Generation Parameters**

### **max_length**

Controls maximum output length.

```python
max_length=80
```

---

### **temperature**

Controls randomness.

- Lower → More predictable
- Higher → More creative

Example:

```python
temperature=0.8
```

---

### **top_k**

Limits token selection to the top K candidates.

```python
top_k=50
```

---

### **top_p**

Uses nucleus sampling.

```python
top_p=0.95
```

The model chooses from the smallest set of likely tokens whose cumulative probability reaches 95%.
