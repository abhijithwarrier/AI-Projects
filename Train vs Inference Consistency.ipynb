**PYTHON SCRIPT TO _ENSURE CONSISTENT PREPROCESSING BETWEEN TRAINING AND INFERENCE IN ML SYSTEMS_. 🧠🔄🏭**

This script demonstrates one of the most critical concepts in production ML: ensuring that the **same preprocessing steps applied during training are also applied during inference**.

Inconsistent pipelines are a major cause of real-world ML failures.

---

## **📦 Import Required Libraries**

We use scikit-learn utilities for preprocessing, modeling, and evaluation.

```python
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

---

## **🧩 Load Dataset**

```python
data = load_wine()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🚨 ❌ WRONG APPROACH – Inconsistent Inference**

Here we scale training data but forget to apply the same transformation during inference.

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

model = LogisticRegression(max_iter=5000)
model.fit(X_train_scaled, y_train)

# ❌ Inference WITHOUT scaling
y_pred_wrong = model.predict(X_test)

print("❌ Accuracy with inconsistent pipeline:", accuracy_score(y_test, y_pred_wrong))
```

👉 Problem:

- Model trained on scaled data
- Predictions made on unscaled data
- Results become unreliable

---

## **✅ CORRECT APPROACH – Manual Consistency**

We ensure the same transformation is applied at inference time.

```python
X_test_scaled = scaler.transform(X_test)

y_pred_correct = model.predict(X_test_scaled)

print("✅ Accuracy with consistent preprocessing:", accuracy_score(y_test, y_pred_correct))
```

---

## **🚀 BEST PRACTICE – Using Pipeline**

The safest way to guarantee consistency is using a pipeline.

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=5000))
])

pipeline.fit(X_train, y_train)

y_pred_pipeline = pipeline.predict(X_test)

print("🚀 Pipeline Accuracy:", accuracy_score(y_test, y_pred_pipeline))
```

The pipeline automatically ensures:

- training transformations = inference transformations
- no manual mistakes
- production-safe workflow

---

## **🔍 Why This Matters in Production**

In real-world systems:

- data is processed separately from training
- inference happens in APIs or batch jobs
- manual preprocessing is error-prone

If training and inference pipelines differ:

- model performance drops
- predictions become inconsistent
- systems fail silently

---

## **🧠 Key Takeaways**

1. Training and inference pipelines must be identical.
2. Applying transformations only during training is a common mistake.
3. Inconsistent preprocessing leads to incorrect predictions.
4. Pipelines ensure automatic consistency.
5. A critical concept for production-ready ML systems.

---

## **Conclusion**

Train vs inference consistency is one of the most important principles in production machine learning. Even a perfectly trained model can fail if preprocessing steps are not applied consistently. By using pipelines, we ensure reliable and repeatable ML workflows.
