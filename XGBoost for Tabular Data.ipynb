**PYTHON SCRIPT TO _TRAIN A HIGH-PERFORMANCE XGBOOST MODEL FOR TABULAR CLASSIFICATION TASKS_. 🧠🚀📊**

This script demonstrates how to use **XGBoost**, one of the most powerful gradient boosting algorithms, for **tabular data classification**. XGBoost is widely used in real-world ML systems because of its strong performance, efficiency, and flexibility.

---

## **📦 Import Required Libraries**

We use pandas for tabular data handling, scikit-learn for dataset splitting and evaluation, and XGBoost for modeling.

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

from xgboost import XGBClassifier
```

---

## **🧩 Load the Dataset**

We use a structured classification dataset that works well for tabular ML.

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

---

## **✂️ Split Data into Train and Test Sets**

We keep a separate test set to evaluate the model fairly.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

---

## **🤖 Initialise the XGBoost Model**

We configure the classifier with a practical starting set of hyper-parameters.

```python
xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric="logloss"
)
```

These parameters control:

- number of boosting rounds
- learning speed
- tree complexity
- row and column sampling

---

## **🚀 Train the Model**

```python
xgb_model.fit(X_train, y_train)
```

XGBoost builds trees sequentially, with each new tree correcting errors made by previous ones.

---

## **📊 Generate Predictions**

We make both class predictions and probability predictions.

```python
y_pred = xgb_model.predict(X_test)
y_probs = xgb_model.predict_proba(X_test)[:, 1]
```

This allows us to evaluate both classification accuracy and probability quality.

---

## **✅ Evaluate Model Performance**

We use common classification metrics to understand performance.

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_probs))
```

This gives us:

- overall accuracy
- precision / recall / F1-score
- ROC-AUC for class separation quality
