Large Language Models don't simply answer questions—they follow **instructions**. The way you write those instructions, known as **prompts**, has a significant impact on the quality of the AI's responses.

Two of the most common prompting techniques are **Zero-shot Prompting** and **Few-shot Prompting**. Zero-shot prompting relies entirely on the model's existing knowledge, while Few-shot prompting provides examples to guide the model toward the desired output.

In this project, we'll explore both approaches and compare their outputs using a Hugging Face instruction-following language model.

We'll learn how to:

- Perform Zero-shot prompting
- Perform Few-shot prompting
- Compare responses from both techniques
- Understand when each approach should be used
- Improve prompt quality for LLM applications

## 📦 Import Required Libraries

````python
from transformers import pipeline
````

We'll use the Hugging Face `pipeline()` API to quickly load an instruction-following language model.

---

## 🤖 Load an Instruction-Following Model

````python
generator = pipeline(
    "text-generation",
    model="google/gemma-2b-it"
)
````

Instruction-tuned models are specifically trained to understand prompts and generate useful, conversational responses.

---

## 📝 Create a Zero-shot Prompt

````python
zero_shot_prompt = """
Translate the following sentence into French.

Sentence:
Python is an amazing programming language.
"""
````

A Zero-shot prompt gives the model only the task description without providing any examples.

---

## 🚀 Generate Zero-shot Response

````python
zero_shot_response = generator(
    zero_shot_prompt,
    max_new_tokens=60,
    do_sample=False
)

print(zero_shot_response[0]["generated_text"])
````

The model relies entirely on its pretrained knowledge to complete the task.

---

## 📝 Create a Few-shot Prompt

````python
few_shot_prompt = """
English: Good Morning
French: Bonjour

English: Thank You
French: Merci

English: Good Night
French: Bonne Nuit

English: Python is an amazing programming language.
French:
"""
````

Few-shot prompting provides a few examples before asking the model to perform the same task.

---

## 🚀 Generate Few-shot Response

````python
few_shot_response = generator(
    few_shot_prompt,
    max_new_tokens=60,
    do_sample=False
)

print(few_shot_response[0]["generated_text"])
````

The examples help the model better understand the expected output format and style.

---

## 📊 Compare Another Example

````python
tasks = [
    "Classify the sentiment: I absolutely love this movie.",
    "Summarize: Python is a high-level programming language designed for readability.",
    "Explain recursion in one sentence."
]

for task in tasks:
    print("\n" + "=" * 60)
    print(task)

    result = generator(
        task,
        max_new_tokens=80,
        do_sample=False
    )

    print(result[0]["generated_text"])
````

Zero-shot prompting performs surprisingly well for many common NLP tasks because modern instruction-tuned models have already learned to follow natural language instructions.

---

## 🔄 Few-shot Prompt for Classification

````python
classification_prompt = """
Review: This phone is fantastic.
Sentiment: Positive

Review: The service was terrible.
Sentiment: Negative

Review: I am extremely happy with this purchase.
Sentiment:
"""

result = generator(
    classification_prompt,
    max_new_tokens=20,
    do_sample=False
)

print(result[0]["generated_text"])
````

Providing a few labeled examples often improves consistency for classification tasks.

## 📈 Zero-shot vs Few-shot Workflow

```
User Task
     │
     ▼
Write Prompt
     │
     ├───────────────┐
     │               │
     ▼               ▼
Zero-shot        Few-shot
(No Examples)   (Few Examples)
     │               │
     ▼               ▼
LLM Generates Response
     │
     ▼
Evaluate Output
```

This illustrates how both prompting strategies guide the model toward generating responses, with Few-shot prompting offering additional context through examples.

## 🌍 Real-World Applications

Zero-shot prompting is commonly used for:

- General question answering
- Text summarization
- Content generation
- Translation
- Brainstorming ideas

Few-shot prompting is widely used for:

- Text classification
- Structured data extraction
- Named Entity Recognition (NER)
- Format-specific content generation
- Consistent response formatting
- Domain-specific AI assistants

## 🧠 Key Takeaways

- Zero-shot prompting provides only the task description.
- Few-shot prompting includes example inputs and outputs.
- Examples help guide the model toward the desired response style.
- Few-shot prompting often improves consistency for structured tasks.
- Choosing the right prompting technique can significantly improve LLM performance.
